# Laços com while

Agora que sabemos vários loops, podemos brincar com várias tarefas repetitivas. Vamos supor que você queira criar um chatbot, assim como o chatbot abaixo (para terminar a conversa é só escrever "tchau").

Em muitos casos, como esse, em que sabemos que nossa contagem será sempre de i até 0, podemos transformar o nosso laço *while* em um for:

````python
for i in range(10, 0, -1):
  print(str(i)+'!')
print('Feliz ano novo!')
````

Porém, uma coisa a se notar é que no while precisamos deixar explícito o *incremento* da nossa variável. Por isso, precisamos da linha abaixo:

```
i = i - 1
```

O que acontece se nos esquecermos dessa linha?

````python
#@title Chatbot (código ultra secreto)
nome = input('Oi! Qual é o seu nome?\n')

while nome != 'tchau':
  print('Que legal! O que é ' + nome + '?')
  nome = input()
````

Legal, vamos estruturar nosso for! Quantas vezes ele vai rodar?

Temos um pequeno problema: nós *não sabemos*! Neste contexto, o usuário pode escrever um número indefinido de mensagens, e não conseguimos estruturar um for sem saber o número exato de vezes que ele deve rodar, seja com um número na função range, seja com o tamanho de uma lista.

Para esses casos, quando não sabemos o número definido de vezes que nosso loop deverá rodar, temos o loop *while*. Veja abaixo um pequeno exemplo:

````python
i = 10
while i > 0:
  print(str(i) + '!')
  i = i - 1
print('Feliz ano novo!')
````

````python
i = 10
while i > 0:
  print(str(i) + '!')
print('Feliz ano novo!')
````

Entramos aqui num problema que nunca tivemos com os laços for: o *loop infinito*. Como o while roda um número indefinido de vezes, se ele nunca chegar na sua *condição de parada* (a condição do while), nós vamos executar o código no while infinitamente!

Obs.: Em alguns casos, isso é até o que queremos, mas isso fica pra outra aula :)

````python
# Evite um loop infinito!
i = 0
while i < 11:
  if i%2 == 0:
    print('bem me quer')
  else:
    print('mal me quer')
  i = i + 1
````

````python
# Evite um loop infinito!
finalizado = False
while not finalizado:
  print('ta finalizado?')
  resposta = input()
  if resposta == 'sim':
    finalizado = True
  else:
    finalizado = False
````

````python
# Evite um loop infinito!
a = 0
while a < 10:
  print('0 é menor que 10')
  a += 1
print('0 ainda é menor que 10 mas o loop acabou')
````
