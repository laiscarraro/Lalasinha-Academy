## Variáveis

Explicando o que são variáveis.

## Quais nomes são possíveis?

Existem algumas regras para nomeação de variáveis:

- Não podem ser **palavras reservadas** da liguagem (que têm significados predefinidos)
- Não podem começar com números (pode até conter, mas não no primeiro caracter)
- Não podem conter .,#$%&!"@()\[\]*{}\;:?\'|\/\ pois todos esses caracteres têm outros significados
- Até podem conter acentuação (´, `, ^, ~) ou 'ç', mas melhor não (tem potencial de dar ruim)
- Podemos chamar nossas variáveis com mais de uma palavra (geralmente 2 ou 3, pois mais que isso fica muito confuso), mas não podemos separá-las por espaço.

Algumas alternativas comuns para separar as palavras sem espaço:
- **Camel case**: a primeira letra de cada palavra, *com exceção da primeira*, é escrita em letra maiúscula. Exemplos:
  - mediaPonderada
  - numDeEntradas
  - primeiraPalavra

- **Snake case**: as palavras são separadas por underscores (_) ao invés de espaços. O mais comum em python! Exemplos:
  - numero_casas
  - idade_pessoa
  - tipo_lista

Outros nomes válidos de variáveis:
- numeroA
- numero1
- t1
- teste_1234
- tesTe12A3
- lalalalalalalalalalalalalalalalalala

## Dicas
- Tente não deixar suas variáveis com um nome muito grande (tipo essa última, ou com o tamanho de uma frase, por exemplo), pois vai ser chato escrever o nome dela várias vezes depois.
- Tente não deixar os nomes muito pequenos também (a, i, j, k, t1, t2, t3, etc.) porque com 3 ou mais variáveis você com certeza vai se perder.
- Para facilitar, dê nomes intuitivos e sucintos para suas variáveis, assim vai ser muito mais fácil de lembrar e você não vai precisar ficar voltando na definição de variávies toda hora.
- Outra dica é manter um padrão na definição de variáveis. Se usou snake_case em uma, não use camelCase em outras. Evite misturar cases.
- A última dica é definir suas variáveis num mesmo lugar (quando fizer sentido) do código. É sempre bom ter um "cantinho das variáveis" para saber onde voltar se precisar.
