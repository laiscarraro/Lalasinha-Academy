# Interação com o usuário - função *input()*

Se você já usou algum programa, já percebeu que é muito comum o programa pedir dados do usuário, sem que ele tenha que inserí-los diretamente no código. Exemplos disso são formulários, textos de redes sociais, ou até mesmo os códigos que nós escrevemos no colab! Mas como fazer isso em Python?

Temos a seguinte função:

```
input()
```

O que ela faz é simplemente pegar um string de tamanho qualquer que o usuário digitar. Teste na célula abaixo!




Mas cuidado! Essa função retorna apenas STRINGS, ou seja, mesmo que você peça para o usuário escrever um número, o retorno da função vai ser uma string. Por exemplo:

````python
cateto_a = input('digite o cateto a: ')
cateto_b = input('digite o cateto b: ')
print('A hipotenusa é ' + str(cateto_a**2 + cateto_b**2))
````

Podemos corrigir usando uma conversão de tipo para inteiro:

````python
cateto_a = int(input('digite o cateto a: '))
cateto_b = int(input('digite o cateto b: '))
print('A hipotenusa é ' + str(cateto_a**2 + cateto_b**2))
````

````python
input()
````

Legal, né? Podemos, também, armazená-la numa variável para poder manipular esse input de diversas formas:

````python
nome = input()
print('Oi, ' + nome + '!')
````

Já viu como nos formulários existem indicações do que você deve preencher em cada campo? Podemos fazer isso aqui também! É só passar para a função input uma string com a mensagem que você quer mostrar para o seu usuário :)

````python
nome = input('Digite seu nome: ')
altura = input('Digite sua altura: ')

print('Seu nome é ' + nome + ' e você mede ' + altura + '.')
````
