# Break e Continue

Vamos supor que queremos construir um chatbot que pare de conversar quando o usuário escreve "tchau". Podemos fazer isso de algumas formas diferentes. Primeiro, vamos ver como faríamos apenas com as ferramentas que já temos:

```python
resp = ''
while resp != 'tchau':
  print('Oi, eu sou um chatbot, e vc?')
  resp = input()
print('Tchau!')
```

Mas podemos também usar uma palavra especial chamada _break_. O _break_ faz com que o loop termine imediatamente, sem executar mais nada dentro do loop (mas abaixo ainda executa!):

```python
resp = ''
while True:
  print('Oi, eu sou um chatbot, e vc?')
  resp = input()]
  if resp == 'tchau':
    break
print('Tchau!')
```

(perceba como ele ainda imprime o 'tchau' do final)

Vamos supor que, ainda nesse mesmo chatbot, queremos que ele "ignore" algumas respostas do usuário. Por exemplo, digamos que queremos que o chatbot ignore a palavra "sim". Como faríamos isso? Podemos até tentar deixar o elif sem nada, mas será que vai dar certo?

```python
resp = ''
while True:
  print('Oi, eu sou um chatbot, e vc?')
  resp = input()
  if resp == 'tchau':
    break
  elif resp == 'sim':
    # não fazer nada
#print('Tchau!')
```

Ih! dá um erro. Então como será que a gente faz pra que ele ignore alguma parte do código? Usamos a palavra _continue_. Assim como no _break_, tudo depois do _continue_ e dentro do loop é ignorado. No entanto, ao invés de sair do loop, o _continue_ volta para o começo do laço de novo.

OBS.: Vamos adicionar uma resposta "imediata" do chatbot ("Legal!") para que dê pra notar quanto ele ignorar.

```python
resp = ''
while True:
  print('Oi, eu sou um chatbot, e vc?')
  resp = input()
  if resp == 'tchau':
    break
  elif resp == 'sim':
    continue
  print('Legal!')
print('Tchau!')
```

OBS.: Usamos essas palavrinhas especiais só no while, mas dá pra usar no for também :)
